```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_excel('Ghy_Attedence.xlsx')
```

```python
data
```

```python
# Parse the date using the desired format string
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')
```

```python
# Set Date as the index
data = data.set_index(['Date'])
```

```python
data.head()
```

```python
# Group by Employee ID and Date
grouped_df = data.groupby(['Emp Name' , 'Emp Code', 'Date'])

# Extract the first in time and last out time
first_in_time = grouped_df['Time'].min()
last_out_time = grouped_df['Time'].max()

# Combine the first in time and last out time into a new dataframe
result_df = pd.DataFrame({'First In Time': first_in_time, 'Last Out Time': last_out_time})

# Print the result
print(result_df)
```

```python
result_df.head(30)
```

```python
# Save the result to a CSV file
result_df.to_csv('attendance_data.csv', index=False)
```
